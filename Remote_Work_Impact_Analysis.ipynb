```python
### Question 2: Is there a correlation between weekly telework hours and age?

telework_hours_data = table_3[table_3['Characteristic'].str.contains('years|Men|Women')].copy()

telework_hours_data.reset_index(drop=True, inplace=True)

# Filter age groups
age_telework = telework_hours_data[telework_hours_data['Characteristic'].str.contains('years')].copy()

# Exclude combined categories
age_telework = age_telework[~age_telework['Characteristic'].isin(['Race and Hispanic or Latino ethnicity',
                                                                   'Disability status', 'Foreign born status',
                                                                   'Marital status', 'Presence of children',
                                                                   'Educational attainment'])]

# Convert age groups to numerical for correlation (use midpoint of age ranges)
def get_age_midpoint(age_group):
    # Check if the entry relates to an age range
    if 'years' not in age_group:
        return np.nan  # Skip non-age entries like "Men, 16 years and over"
    elif 'and over' in age_group:  # Handle "55 years and over"
        try:
            return int(age_group.split()[0])
        except ValueError:
            return np.nan
    elif 'to' in age_group:  # Handle ranges like "25 to 34 years"
        try:
            parts = age_group.split(' to ')
            low, high = int(parts[0]), int(parts[1].split()[0])
            return (low + high) / 2
        except (ValueError, IndexError):
            return np.nan
    else:
        return np.nan

# Apply the function to calculate Age Midpoint
age_telework['Age_Midpoint'] = age_telework['Characteristic'].apply(get_age_midpoint)

# Filter gender categories
gender_telework = telework_hours_data[telework_hours_data['Characteristic'].isin(['Men, 16 years and over', 'Women, 16 years and over'])].copy()

# Rename for clarity
gender_telework.rename(columns={
    'Characteristic': 'Gender',
    'Avg_Weekly_Hours_Teleworked': 'Avg_Weekly_Hours_Teleworked'
}, inplace=True)

# Display gender telework hours
display(gender_telework[['Gender', 'Avg_Weekly_Hours_Teleworked']])

# Drop rows with NaN in Age_Midpoint or Avg_Weekly_Hours_Teleworked
age_corr_data = age_telework.dropna(subset=['Age_Midpoint', 'Avg_Weekly_Hours_Teleworked'])

# Calculate Pearson correlation
age_correlation = age_corr_data['Age_Midpoint'].corr(age_corr_data['Avg_Weekly_Hours_Teleworked'])
print(f"Pearson Correlation between Age and Avg Weekly Telework Hours: {age_correlation:.2f}")



plt.figure(figsize=(10, 6))
sns.scatterplot(data=age_corr_data, x='Age_Midpoint', y='Avg_Weekly_Hours_Teleworked')
sns.regplot(data=age_corr_data, x='Age_Midpoint', y='Avg_Weekly_Hours_Teleworked', scatter=False, color='red')
plt.title('Correlation Between Age and Average Weekly Telework Hours')
plt.xlabel('Age (Midpoint)')
plt.ylabel('Average Weekly Telework Hours')
plt.tight_layout()
plt.show()
```

The scatter plot visualizes the correlation between age (using midpoint values of age ranges) and average weekly telework hours. The red regression line shows a positive trend indicating that as age increases, average weekly telework hours also tend to increase. The calculated Pearson correlation coefficient of 0.55 suggests a moderate positive correlation between age and telework hours. This implies that older age groups generally telework more hours per week compared to younger groups. However, the moderate correlation also indicates variability meaning age is not the sole predictor of telework hours.

Additionally, the gender-specific data reveals that women, on average, telework more hours per week than men among those aged 16 and over. Specifically, women average 7.1 hours, whereas men average 5.6 hours. This difference could be influenced by various factors such as job roles, family responsibilities, or workplace flexibility.

#  Impact of Occupation on Telework

```python
# ---------------------------
# Research Question 1: Do specific job categories (e.g., managerial, administrative) show higher levels of telework?
# ---------------------------

# Extract Occupation Data from Table 2
occupation_start_idx = table_2[table_2['Characteristic'] == 'Occupation'].index
if not occupation_start_idx.empty:
    occupation_start_idx = occupation_start_idx[0] + 1
    occupation_data_2 = table_2.iloc[occupation_start_idx:].copy()
else:
    occupation_keywords = ['Management', 'Professional', 'Administrative', 'Operations', 'Engineering', 'Science', 'Services', 'Sales', 'Support', 'Construction', 'Manufacturing', 'Transportation', 'Finance', 'Real Estate', 'Healthcare']
    occupation_data_2 = table_2[table_2['Characteristic'].str.contains('|'.join(occupation_keywords), case=False, na=False)].copy()

# Drop non-occupation rows
occupation_data_2 = occupation_data_2[~occupation_data_2['Characteristic'].str.contains('workers|Total|Full-time|Part-time', case=False, na=False)]

# Reset index
occupation_data_2.reset_index(drop=True, inplace=True)

# Convert columns to numeric
numeric_cols = [
    'Total_persons_at_work',
    'Teleworked_Total',
    'Teleworked_Some_Hours',
    'Teleworked_All_Hours',
    'Did_Not_Telework',
    'Percent_Distribution_Total',
    'Percent_Distribution_Teleworked_Total',
    'Percent_Distribution_Some_Hours',
    'Percent_Distribution_All_Hours',
    'Percent_Distribution_Did_Not_Telework'
]
for col in numeric_cols:
    if col in occupation_data_2.columns:
        occupation_data_2[col] = pd.to_numeric(occupation_data_2[col], errors='coerce')

# Drop rows with missing critical data
occupation_data_2.dropna(subset=['Teleworked_Total', 'Total_persons_at_work'], inplace=True)

# Calculate Telework Rate (%)
occupation_data_2['Telework_Rate (%)'] = (occupation_data_2['Teleworked_Total'] / occupation_data_2['Total_persons_at_work']) * 100

# Calculate Average Weekly Telework Hours
occupation_data_2['Avg_Weekly_Hours_Teleworked'] = (
    (occupation_data_2['Teleworked_Some_Hours'] + occupation_data_2['Teleworked_All_Hours']) /
    occupation_data_2['Teleworked_Total']
)

# Visualization: Telework Rates by Job Category
plt.figure(figsize=(14, 10))
sns.barplot(
    data=occupation_data_2.sort_values('Telework_Rate (%)', ascending=False),
    x='Telework_Rate (%)',
    y='Characteristic',
    palette='coolwarm'
)
plt.title('Telework Participation Rate by Job Category')
plt.xlabel('Telework Rate (%)')
plt.ylabel('Job Category')
plt.tight_layout()
plt.show()


```

The graph shows telework participation rates by job category with a clear gradient from high to low telework rates. Occupations in Computer and Mathematical, Finance and Insurance, Business and Financial Operations, and Legal fields have the highest telework rates, often exceeding 50%, due to their knowledge-based, less location-dependent nature. Mid-range telework participation is seen in Arts and Design, Management, and Education, where some on-site presence is needed. At the bottom are occupations like Food Preparation, Transportation, Construction, and Farming, with telework rates below 10%, as these roles inherently require physical presence. This visualization effectively highlights how job functions correlate with telework feasibility.

```python
# ---------------------------
# Research Question 2: How does industry impact telework participation and hours?
# ---------------------------

# Extract Industry Data from Table 4
industry_start_idx = table_4[table_4['Characteristic'] == 'Industry'].index
if not industry_start_idx.empty:
    industry_start_idx = industry_start_idx[0] + 1
    industry_data_4 = table_4.iloc[industry_start_idx:].copy()
else:
    industry_keywords = ['Agriculture', 'Nonagricultural', 'Mining', 'Construction', 'Manufacturing', 'Wholesale', 'Retail', 'Transportation', 'Utilities', 'Information', 'Financial', 'Real estate', 'Finance']
    industry_data_4 = table_4[table_4['Characteristic'].str.contains('|'.join(industry_keywords), case=False, na=False)].copy()

# Drop non-industry rows
industry_data_4 = industry_data_4[~industry_data_4['Characteristic'].str.contains('workers|Total|Full-time|Part-time|Occupation', case=False, na=False)]

# Reset index
industry_data_4.reset_index(drop=True, inplace=True)

# Convert columns to numeric
numeric_cols_industry = [
    'Total_Persons_at_Work_(thousands)',
    'Avg_Weekly_Hours_Teleworked',
    'Avg_Weekly_Hours_Worked',
    'Telework_Hours_as_Percent_of_Worked_Hours',
    'Teleworked_Persons_(thousands)',
    'Teleworked_Avg_Weekly_Hours',
    'Teleworked_Avg_Weekly_Worked_Hours',
    'Teleworked_Hours_as_Percent_of_Worked_Hours'
]
for col in numeric_cols_industry:
    if col in industry_data_4.columns:
        industry_data_4[col] = pd.to_numeric(industry_data_4[col], errors='coerce')

# Drop rows with missing critical data
industry_data_4.dropna(subset=['Teleworked_Persons_(thousands)', 'Total_Persons_at_Work_(thousands)'], inplace=True)

# Calculate Telework Rate (%)
industry_data_4['Teleworked_Total'] = industry_data_4['Teleworked_Persons_(thousands)'] * 1000
industry_data_4['Total_Persons_at_Work'] = industry_data_4['Total_Persons_at_Work_(thousands)'] * 1000
industry_data_4['Telework_Rate (%)'] = (industry_data_4['Teleworked_Total'] / industry_data_4['Total_Persons_at_Work']) * 100

# Calculate Average Weekly Telework Hours
industry_data_4['Teleworked_Avg_Weekly_Hours'] = pd.to_numeric(industry_data_4['Teleworked_Avg_Weekly_Hours'], errors='coerce')
industry_data_4['Avg_Weekly_Hours_Teleworked'] = industry_data_4['Teleworked_Avg_Weekly_Hours']

# Handle missing values by calculating average from 'Teleworked_Some_Hours' and 'Teleworked_All_Hours' if needed
if 'Teleworked_Some_Hours' in industry_data_4.columns and 'Teleworked_All_Hours' in industry_data_4.columns:
    missing_hours_industry = industry_data_4['Avg_Weekly_Hours_Teleworked'].isnull()
    if missing_hours_industry.any():
        industry_data_4.loc[missing_hours_industry, 'Avg_Weekly_Hours_Teleworked'] = (
            (industry_data_4.loc[missing_hours_industry, 'Teleworked_Some_Hours'] + industry_data_4.loc[missing_hours_industry, 'Teleworked_All_Hours']) /
            industry_data_4.loc[missing_hours_industry, 'Teleworked_Total']
        )

# Visualization: Telework Rates by Industry
plt.figure(figsize=(14, 10))
sns.barplot(
    data=industry_data_4.sort_values('Telework_Rate (%)', ascending=False),
    x='Telework_Rate (%)',
    y='Characteristic',
    palette='viridis'
)
plt.title('Telework Participation Rate by Industry')
plt.xlabel('Telework Rate (%)')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()
```

The graph displays telework participation rates across various industries. Industries such as Finance and Insurance, Professional and Technical Services, Financial Activities, and Information show the highest telework participation, with rates over 50% suggesting that these knowledge-based sectors are well-suited for remote work. Mid-range telework participation is seen in Public Administration, Social Assistance, and Manufacturinn where certain roles may accommodate remote work but others require physical presence. Accommodation and Food Services, Transportation, and Leisure and Hospitality have the lowest telework rates, under 10% as these industries largely rely on in-person service. This highlights how the nature of work within each industry affects its adaptability to telework.

```python
# Visualization: Average Weekly Telework Hours by Industry
plt.figure(figsize=(14, 10))
sns.barplot(
    data=industry_data_4.sort_values('Avg_Weekly_Hours_Teleworked', ascending=False),
    x='Avg_Weekly_Hours_Teleworked',
    y='Characteristic',
    palette='plasma'
)
plt.title('Average Weekly Telework Hours by Industry')
plt.xlabel('Average Weekly Telework Hours')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()
```

This graph visualizes the Average Weekly Telework Hours across different industries. The Agriculture and Related Industries and Professional and Technical Services sectors exhibit the highest average weekly telework hours suggesting a strong adaptability to remote work in these areas. Other sectors with high average telework hours include Professional and Business Services, Finance and Insurance, and Information industries, where job roles are often knowledge-based and flexible for remote work setups. In contrast, industries such as Private Households, Local Government, and Educational Services have the lowest average weekly telework hours, likely due to the necessity for in-person interaction and hands-on tasks in these fields. This distribution highlights the variance in remote work feasibility based on industry characteristics.

# Work Status and Telework Engagement

```python
# ---------------------------
# Research Question 1: Are full-time or part-time employees more likely to telework?
# ---------------------------

table_2 = pd.read_csv(base_path + 'telework_table_2_cleaned.csv', skiprows=1)

table_8 = pd.read_csv(base_path + 'telework_table_8_cleaned.csv', skiprows=1)

# Replace non-breaking spaces with regular spaces in column names
table_2.columns = table_2.columns.str.replace('\xa0', ' ', regex=True)
table_8.columns = table_8.columns.str.replace('\xa0', ' ', regex=True)

# Define complete new column names for Table 2 (11 columns)
columns_table_2 = [
    'Characteristic',                         # 'Usual full- and part-time status'
    'Total_persons_at_work',                  # 'Unnamed: 1'
    'Teleworked_Total',                       # 'Unnamed: 2'
    'Teleworked_Some_Hours',                  # 'Unnamed: 3'
    'Teleworked_All_Hours',                   # 'Unnamed: 4'
    'Did_Not_Telework',                       # 'Unnamed: 5'
    'Percent_Distribution_Total',             # 'Unnamed: 6'
    'Percent_Distribution_Teleworked_Total',  # 'Unnamed: 7'
    'Percent_Distribution_Some_Hours',        # 'Unnamed: 8'
    'Percent_Distribution_All_Hours',         # 'Unnamed: 9'
    'Percent_Distribution_Did_Not_Telework'   # 'Unnamed: 10'
]

# Check if the number of columns matches
if len(columns_table_2) != table_2.shape[1]:
    print("\nError: Number of new column names for Table 2 does not match the number of existing columns.")
else:
    # Assign new column names to Table 2
    table_2.columns = columns_table_2

# Define complete new column names for Table 8 (20 columns)
columns_table_8 = [
    'Characteristic',                             # 'Usual full- and part-time status'
    'Total_Persons_Teleworked_or_Home_Pay',       # 'Unnamed: 1'
    'Up_to_8_hours',                              # 'Unnamed: 2'
    '9_to_16_hours',                              # 'Unnamed: 3'
    '17_to_24_hours',                             # 'Unnamed: 4'
    '25_to_32_hours',                             # 'Unnamed: 5'
    '33_to_39_hours',                             # 'Unnamed: 6'
    'Total',                                      # 'Unnamed: 7'
    '40_hours',                                   # 'Unnamed: 8'
    '41_hours_or_more',                           # 'Unnamed: 9'
    'Avg_Weekly_Hours_Teleworked',                # 'Unnamed: 10'
    'Total_Percent_Distribution',                 # 'Unnamed: 11'
    'Up_to_8_hours_percent',                      # 'Unnamed: 12'
    '9_to_16_hours_percent',                      # 'Unnamed: 13'
    '17_to_24_hours_percent',                     # 'Unnamed: 14'
    '25_to_32_hours_percent',                     # 'Unnamed: 15'
    '33_to_39_hours_percent',                     # 'Unnamed: 16'
    'Total_percent_distribution',                 # 'Unnamed: 17'
    '40_hours_percent',                           # 'Unnamed: 18'
    '41_hours_or_more_percent'                    # 'Unnamed: 19'
]

# Check if the number of columns matches
if len(columns_table_8) != table_8.shape[1]:
    print("\nError: Number of new column names for Table 8 does not match the number of existing columns.")
else:
    # Assign new column names to Table 8
    table_8.columns = columns_table_8


# Define the characteristics to extract (case-insensitive)
work_statuses = ['full-time workers', 'part-time workers']

# Clean the 'Characteristic' column by stripping leading/trailing spaces and converting to lowercase
table_2['Characteristic'] = table_2['Characteristic'].str.strip().str.lower()
table_8['Characteristic'] = table_8['Characteristic'].str.strip().str.lower()

# Remove rows where 'Characteristic' is not in work_statuses
table_2_filtered = table_2[table_2['Characteristic'].isin(work_statuses)].copy()
table_8_filtered = table_8[table_8['Characteristic'].isin(work_statuses)].copy()

# Reset index for both tables
table_2_filtered.reset_index(drop=True, inplace=True)
table_8_filtered.reset_index(drop=True, inplace=True)

# Merge Table 2 and Table 8 on 'Characteristic'
merged_work_status = pd.merge(
    table_2_filtered,
    table_8_filtered,
    on='Characteristic',
    how='inner'
)

# Calculate Telework Rate (%) as (Total_Persons_Teleworked_or_Home_Pay / Total_persons_at_work) * 100
merged_work_status['Telework_Rate (%)'] = (
    merged_work_status['Total_Persons_Teleworked_or_Home_Pay'] /
    merged_work_status['Total_persons_at_work']
) * 100


# ---------------------------
# Step 9: Visualize Telework Participation Rate
# ---------------------------

# Plot Telework Participation Rate by Work Status
plt.figure(figsize=(8, 6))
sns.barplot(data=merged_work_status, x='Characteristic', y='Telework_Rate (%)', palette='viridis')
plt.title('Telework Participation Rate by Work Status')
plt.xlabel('Work Status')
plt.ylabel('Telework Rate (%)')
plt.ylim(0, merged_work_status['Telework_Rate (%)'].max() + 10)  # Adjust based on data range

# Annotate bars with percentage values
for index, row in merged_work_status.iterrows():
    plt.text(
        index,
        row['Telework_Rate (%)'] + 0.5,
        f"{row['Telework_Rate (%)']:.1f}%",
        ha='center',
        va='bottom'
    )

plt.tight_layout()
plt.show()

```

The bar chart illustrates the telework participation rate between full-time and part-time workers revealing a clear disparity. Full-time workers have a higher telework participation rate at 25.1%, compared to just 17.0% for part-time workers. This difference suggests that full-time positions are more conducive to telework, potentially due to the nature of job roles or organizational policies that provide greater flexibility for teleworking in full-time roles. The significant gap also indicates that part-time workers may have fewer opportunities or less access to telework arrangements possibly due to the specific demands or limitations associated with part-time roles. Overall, this visualization underscores a trend where full-time employees are more likely to telework compared to their part-time counterparts.

```python
# ---------------------------
# Part 2: What is the distribution of telework hours among full-time and part-time workers?
# ---------------------------

# Columns related to telework hours (excluding percent columns)
telework_hour_cols = [
    'Characteristic',
    'Up_to_8_hours',
    '9_to_16_hours',
    '17_to_24_hours',
    '25_to_32_hours',
    '33_to_39_hours',
    '40_hours',
    '41_hours_or_more'
]

# Extract and prepare the relevant data for hour breakdown
telework_hours_data = table_8_filtered[telework_hour_cols].copy()

# Ensure all telework hour columns are numeric
telework_hours_data[telework_hour_cols[1:]] = telework_hours_data[telework_hour_cols[1:]].apply(pd.to_numeric, errors='coerce')

# Replace NaN with 0 for plotting
telework_hours_data.fillna(0, inplace=True)

# Reshape the data for easier plotting
telework_hours_melted = telework_hours_data.melt(
    id_vars='Characteristic',
    value_vars=telework_hour_cols[1:],
    var_name='Telework_Hours_Category',
    value_name='Count'
)

# Clean and format the telework hour categories for readability
telework_hours_melted['Telework_Hours_Category'] = (
    telework_hours_melted['Telework_Hours_Category']
    .str.replace('_hours', '', regex=False)
    .str.replace('_', ' ', regex=False)
    .str.title()
)

# Pivot the data for stacked bar plot
pivot_telework_hours = telework_hours_melted.pivot(
    index='Characteristic',
    columns='Telework_Hours_Category',
    values='Count'
).fillna(0)

# Define the order of categories for consistent stacking and filter to match existing columns
hours_order = [
    'Up To 8', '9 To 16', '17 To 24', '25 To 32',
    '33 To 39', '40', '41 Or More'
]
pivot_telework_hours = pivot_telework_hours[[hour for hour in hours_order if hour in pivot_telework_hours.columns]]

# Display the pivoted DataFrame
print("\nPivoted Telework Hours DataFrame:")
print(pivot_telework_hours)

# Plot the stacked bar chart
pivot_telework_hours.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 8),
    colormap='viridis'
)
plt.title('Breakdown of Telework Hours by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Number of Workers')
plt.legend(title='Telework Hours Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

The bar chart shows the breakdown of telework hours for full-time and part-time workers highlighting distinct patterns in telework engagement. Full-time workers dominate in total hours teleworked, with a larger number of individuals clocking high weekly telework hours especially in the categories "40 hours" and "41 or more." This suggests that full-time employees are not only more likely to telework but also tend to telework for extended hours. Part-time workers, in contrast, are concentrated in lower telework hour categories, indicating shorter and less intensive telework engagement. This distribution underscores a significant difference in telework commitment between full-time and part-time employees.

# Conclusion

This analysis examined key factors influencing telework participation, with a focus on demographics, occupation, industry, and employment type. Higher telework rates were found among individuals with advanced degrees, mid-career age groups (25–44), Asian individuals, and women. Knowledge-driven roles such as Computer and Mathematical, Finance, and Legal, as well as industries like Professional Services, Finance, and Information, show the highest telework rate often exceeding 50%. In conclusion, telework engagement varies widely across demographic and occupational factors. Organizations can benefit from tailoring telework policies to meet the distinct needs of these groups, creating a remote work environment that optimizes productivity, flexibility, and employee satisfaction.
