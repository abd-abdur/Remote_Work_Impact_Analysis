In recent years, remote work has become a critical topic impacting how employees manage their work hours especially with respect to telework. This analysis focuses on how teleworking trends vary across demographics, occupations, industries, and work types (full-time vs. part-time). The research aims to provide insights into how these variables influence telework engagement, the distribution of telework hours, and the potential benefits or challenges of remote work in various sectors.

Research Questions


1.   Who Teleworks the Most?
   *  Are certain demographic groups (e.g., age, gender) more likely to telework than others?
   *  Is there a correlation between weekly telework hours and demographics?
  
2.   Does Occupation Influence Telework Hours?
   *  Do specific job categories (e.g., managerial, administrative) show higher levels of telework?
   *  How does industry impact telework participation and hours?
3.   Who Teleworks the Most?
   *  Are full-time or part-time employees more likely to telework?

```python
table_1 = table_1.iloc[:-6]
plt.figure(figsize=(12, 6))
sns.barplot(data=table_1, x='Characteristic', y='Telework_Percentage', palette="viridis")
plt.title('Telework Participation by Age and Gender')
plt.xticks(rotation=90)
plt.ylabel('Telework Percentage')
plt.xlabel('Age Group and Gender')
plt.show()
```

```python
# Plot telework hours as a percentage of total worked hours by demographic group
table_3 = table_3.iloc[:-6]
plt.figure(figsize=(12, 6))
sns.barplot(data=table_3, x='Characteristic', y='Telework_Hours_as_Percent_of_Worked_Hours', palette="magma")
plt.title('Telework Hours as Percentage of Total Work Hours by Age and Gender')
plt.xticks(rotation=90)
plt.ylabel('Telework Hours % of Worked Hours')
plt.xlabel('Age Group and Gender')
plt.show()
```
